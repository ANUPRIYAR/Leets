```python
class Solution:
    def floodFill(self, image: List[List[int]], sr: int, sc: int, color: int) -> List[List[int]]:
        old_color = image[sr][sc]

        def dfs(image, sr, sc, old_color, color):
            if sr < 0 or sr >= len(image) or sc <0 or sc >= len(image[0]):   # Base Condition
                return

            if image[sr][sc] == color:  # Base condition : the pixel should only have old color , 
                return                    # it should neither have current color nor any other color (other than old color)
                                        # Because thats what we have to do , fill old color pixel with new color 
                
            if image[sr][sc] != old_color:     # Base condition    
                return 

            directions = [(1, 0), (-1, 0), (0, 1 ), (0, -1)]
            image[sr][sc] = color    # fill pixel with new color 

            for dx, dy  in directions:
                dfs(image, sr + dx, sc + dy, old_color, color )


        # Call dfs function
        dfs(image, sr, sc, old_color, color)
        return image
```
