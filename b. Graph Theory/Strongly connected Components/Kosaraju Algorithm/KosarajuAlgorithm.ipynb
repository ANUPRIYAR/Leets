## Strongly Connected Components (Kosaraju's Algo
Given a Directed Graph with V vertices (Numbered from 0 to V-1) and E edges, Find the number of strongly connected components in the graph.




Constraints:  
1  V  5000  
0  E  (V*(V-1))  
0  u, v  V-1  
Sum of E over all testcases will not exceed 25*106  

```python
#User function Template for python3

class Solution:
    def dfs(self, node, vis, adj, stack):
        vis[node] = 1
        for it in adj[node]:
            if not vis[it]:
                self.dfs(it, vis, adj, stack)
        stack.append(node)    # run dfs and add the node in stack to track the node with completed dfs traversal
        
    def dfs3(self, node, vis, adjT):
        vis[node] = 1
        for it in adjT[node]:
            if not vis[it]:
                self.dfs3(it, vis, adjT)
        
        
    #Function to find number of strongly connected components in the graph.
    def kosaraju(self, V, adj):
        vis  = [0] * V
        stack = []
        
        for i in range(V):
            if vis[i] == 0:
                self.dfs(i, vis, adj, stack)   # visit each npode and call dfs
                    
        adjT = [[] for _ in range(V)]    # Create Transpose of adj (reversing the directions of edge)
        for i in range(V):
            vis[i] = 0
            for it in adj[i]:
                adjT[it].append(i)   
                
                
        scc = 0
        while stack:
            node = stack.pop()
            if not vis[node]:
                scc += 1
                self.dfs3(node, vis, adjT)     # traverse through stack and run dfs on each node 
                
        return scc
        
```
