Given two integers **L** and **R**, the task is to find the XOR of elements of the range **\[L, R\]**. 

**Examples:**

> **_Input:_** _L = 4, R = 8_ 

> **_Output:_** _8_ 

> _4 ^ 5 ^ 6 ^ 7 ^ 8 = 8_

> **_Input:_** _L = 3, R = 7_ 

> **_Output:_** _3_  

**Naive Approach:** Initialize answer as zero, Traverse all numbers from **L** to **R** and perform XOR of the numbers one by one with the answer. This would take **O(N)** time.

## **XOR of numbers follows a pattern**





```python
# TO FIND THE XOR OF NUMBERS FROM 1 TO N

def find_xor(n):
    if n % 4 == 0:
        return n
    elif n % 4 == 1:
        return 1
    elif n % 4 == 2:
        return n + 1
    elif n % 4 == 3:
        return 0
        
```

```python
# NOW TO FIND THE XOR OF NUMBERS OF GIVEN RANGE L TO H

def xor_range(low, high):
    xor_till_low = find_xor(low-1)
    print(f"xor till {low-1} : {xor_till_low}")
    xor_till_high = find_xor(high)
    print(f"xor till {high} : {xor_till_high}")
    return xor_till_low ^ xor_till_high
```

```python
low = 4
high = 7
print(xor_range(low, high))
```
