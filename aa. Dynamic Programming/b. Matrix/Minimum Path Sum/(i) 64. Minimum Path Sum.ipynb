```python
# Reverse Path Top down 
import math

class Solution:
    def minPathSum(self, grid: List[List[int]]) -> int:
        rows = len(grid)
        cols = len(grid[0])
        dp = [[math.inf for _ in range(cols)] for _ in range(rows)]
        minpath = self.dfs(dp, grid, rows - 1, cols - 1)
        print(dp)
        return minpath

    def dfs(self, dp, grid, r, c):
        rows = len(grid)
        cols = len(grid[0])
        if r == 0  and c == 0:
            return grid[r][c]

        if (r < 0) or ( r >= rows) or (c < 0) or (c >= cols):
            return math.inf

        if dp[r][c] == math.inf:
            moveleft = grid[r][c] + self.dfs(dp, grid, r , c - 1)
            moveup = grid[r][c] +  self.dfs(dp, grid, r - 1 , c)
            dp[r][c] = min(moveleft, moveup)
        return dp[r][c]
```

## Bottom Up Approach 



```python
import math

class Solution:
    def minPathSum(self, grid: List[List[int]]) -> int:
        rows = len(grid)
        cols = len(grid[0])
        dp = [[0 for _ in range(cols)] for _ in range(rows)]
        
        dp[0][0] = grid[0][0]

        for i in range(1, rows):
            dp[i][0] = dp[i-1][0] + grid[i][0]
        
        for j in range(1, cols):
            dp[0][j] = dp[0][j-1] + grid[0][j]

        for i in range(1, rows):
            for j in range(1, cols):
                dp[i][j] = grid[i][j] + min(dp[i][j-1], dp[i-1][j])

        print(dp)
        return dp[rows-1][cols -1 ]
```

```python
grid =[[1,2,3],[4,5,6]]
sol = Solution()
sol.minPathSum(grid)
```
