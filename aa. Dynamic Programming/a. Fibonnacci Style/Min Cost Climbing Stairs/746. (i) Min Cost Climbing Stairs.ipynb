## TLE with recursive solution


```python
from typing import List
class Solution:
    def minCostClimbingStairs(self, cost: List[int]) -> int:
        cost1 = self.helper(cost, 0)
        cost2 = self.helper(cost, 1)
        return min(cost1, cost2)

    def helper(self, cost, cur_idx):
        if cur_idx >= len(cost):
            return 0


        take1step = cost[cur_idx] + self.helper(cost, cur_idx + 1)
        take2step = cost[cur_idx] + self.helper(cost, cur_idx + 2)

        return min(take1step, take2step)
```

## Top down Approach 


```python
class Solution:
    def minCostClimbingStairs(self, cost: List[int]) -> int:
        dp = [-1 for _ in range(len(cost))]
        cost1 = self.helper(dp, cost, 0)
        cost2 = self.helper(dp, cost, 1)
        
        return min( cost1, cost2)

    def helper(self, dp,  cost, cur_idx):
        if cur_idx >= len(cost):
            return 0

        if dp[cur_idx] == -1:
            take1step = cost[cur_idx] + self.helper(dp, cost, cur_idx + 1)
            take2step = cost[cur_idx] + self.helper(dp, cost, cur_idx + 2)
            dp[cur_idx] = min(take1step, take2step)
        return dp[cur_idx]
    
```
