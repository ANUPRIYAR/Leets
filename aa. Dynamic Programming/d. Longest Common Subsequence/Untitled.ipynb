```python
# Using 2 Strings
class Solution:
    def minInsertions(self, s: str) -> int:
        t = s[::-1]
        if s == t:
            return 0
        dp =[[0 for _ in range(len(s)+1)] for _ in range(len(s)+1)]
        n= len(s)

        for i in range(n-1, -1, -1):
            for j in range(n-1, -1,-1):
                if s[i] == t[j]:
                    dp[i][j] = 1 + dp[i+1][j+1]
                else:
                    c1 = dp[i+1][j]
                    c2 = dp[i][j+1]
                    dp[i][j] = max(c1, c2)
        return len(s) - dp[0][0] 
```
