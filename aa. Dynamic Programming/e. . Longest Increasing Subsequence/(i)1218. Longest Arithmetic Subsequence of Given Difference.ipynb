```python
class Solution:
    def longestSubsequence(self, arr: List[int], difference: int) -> int:
        n  = len(arr)
        dp = [1 for _ in range(len(arr))]
        # dp[0] = 1

        maxlength = 1
        for i in range(n):
            for j in range(i):
                if arr[i] - arr[j] == difference and dp[i] <= dp[j]:
                    dp[i]  = 1 + dp[j]
                    maxlength = max(maxlength, dp[i])

        return maxlength
```
