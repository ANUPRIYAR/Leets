```python
# Recursive 
from typing import List
class Solution:
    def lengthOfLIS(self, nums: List[int]) -> int:
        return self.helper(nums, 0, -1)

    def helper(self, nums, cur_idx, prev):
        if cur_idx == len(nums):
            return 0
        c1 = 0
        if prev == -1 or nums[cur_idx] > nums[prev]:
            c1 =  1 + self.helper(nums, cur_idx + 1, cur_idx)

        c2 = self.helper(nums, cur_idx + 1, prev)
        return max(c1, c2)
```

## Top down


```python
class Solution:
    def lengthOfLIS(self, nums: List[int]) -> int:
        dp = [[-1 for _ in range(len(nums))] for _ in range(len(nums))]
        return self.helper(dp, nums, 0, -1)

    def helper(self, dp, nums, cur_idx, prev):
        if cur_idx == len(nums):
            return 0
        c1 = 0
        if dp[cur_idx][prev] == -1:
            if prev == -1 or nums[cur_idx] > nums[prev]:
                c1 =  1 + self.helper(dp, nums, cur_idx + 1, cur_idx)

            c2 = self.helper(dp, nums, cur_idx + 1, prev)
            dp[cur_idx][prev] = max(c1, c2)
        return dp[cur_idx][prev]
```

```python
class Solution:
    def lengthOfLIS(self, nums: List[int]) -> int:
        n = len(nums)
        dp = [0 for _ in range(len(nums))] 

        for i in range(n):
            dp[i] = 1

        maxlength = 1
        for i in range(1, n):
            for j in range(i):
                if nums[i] > nums[j]  and dp[i] <= dp[j] :  # getting max
                    dp[i] = dp[j] + 1
                    maxlength = max(maxlength, dp[i])
        return maxlength
```
