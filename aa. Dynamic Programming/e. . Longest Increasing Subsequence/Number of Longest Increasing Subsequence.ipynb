```python
from typing import List
class Solution:
    def findNumberOfLIS(self, nums: List[int]) -> int:
        dp = {}
        n = len(nums)

        def helper(idx):
            if idx in dp:
                return dp[idx]
            maxlength = 1
            count = 1

            for i in range(idx):
                if nums[i] < nums[idx]:
                    length, cnt = helper(i)
                    if length + 1 > maxlength:
                        maxlength = length + 1
                        count = cnt
                    elif length + 1 == maxlength:
                        count += cnt
            dp[idx] = (maxlength, count)
            return dp[idx] 

        maxofall = 0
        total_count = 0

        for i in range(n):
            length, count = helper(i)
            if length > maxofall:
                maxofall= length
                total_count = count
            elif length == maxofall:
                total_count += count

        return total_count
```

```python
class Solution:
    def findNumberOfLIS(self, nums: List[int]) -> int:
        n = len(nums)
        dp = [1] * n
        counts = [1] * n

        for i in range(n):
            for j in range(i):
                if nums[i] > nums[j]:
                    if dp[i] < dp[j] + 1:
                        dp[i] = dp[j] + 1
                        counts[i] = counts[j]
                    elif dp[j] + 1 == dp[i]:
                        counts[i] += counts[j]

        maxlength = max(dp)
        return sum(c for l, c in zip(dp, counts) if l == maxlength)
```
