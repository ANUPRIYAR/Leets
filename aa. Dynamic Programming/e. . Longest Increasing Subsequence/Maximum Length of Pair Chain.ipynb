```python
from typing import List

class Solution:
    def findLongestChain(self, pairs: List[List[int]]) -> int:
        pairs.sort(key = lambda x: x[1])
        dp = [[-1 for _ in range(len(pairs))] for _ in range(len(pairs))]
        return self.helper(dp, pairs, 0, -1)

    def helper(self, dp, pairs, idx, prev):
        if idx == len(pairs):
            return 0

        c1, c2 = 0, 0
        if dp[idx][prev] == -1:
            if prev == -1 or pairs[prev][1] < pairs[idx][0]:
                c1 = 1 + self.helper(dp, pairs, idx + 1, idx)
                c2 = self.helper(dp, pairs, idx + 1, prev)
            
            c3 = self.helper(dp, pairs, idx + 1, prev)
            dp[idx][prev] = max(c3, max(c1, c2))
        return dp[idx][prev]
```

```python
# O(n) solution
from typing import List
class Solution:
    def findLongestChain(self, pairs: List[List[int]]) -> int:
        pairs.sort(key = lambda x: x[1])
        count = 0
        curr = float('-inf')
        for start, end in pairs:
            if start > curr:
                curr = end
                count += 1
            elif start < curr and end < curr:
                curr = end
        return count
```
