## Top down approach 


```python
class Solution:
    def numDistinct(self, s: str, t: str) -> int:
        dp = [[-1 for _ in range(len(t))] for _  in range(len(s))]
        return self.helper(dp, s, t, 0, 0)

    def helper(self, dp, s, t, i, j):
        if j == len(t):
            return 1 

        if i == len(s):
            return 0
        
        if dp[i][j] == -1:
            if s[i] == t[j]:
                dp[i][j] =  self.helper(dp, s, t, i+1, j+1) + self.helper(dp, s, t, i+1, j)
            else:
                dp[i][j] = self.helper(dp, s, t, i+1, j)
        return dp[i][j]
        
```

## Bottom Up


```python
class Solution:
    def numDistinct(self, s: str, t: str) -> int:
        slen = len(s)
        tlen = len(t)
        dp = [[0 for _ in range(len(t) + 1)] for _  in range(len(s) + 1 )]

        for i in range(slen + 1): # we are keeping it slen +1 because the entire column should be 1 .
            dp[i][tlen] = 1               #And if we are putting just 1, it doesnt matter which direction we take this particular for loop (slen +1)


        for i in range(slen-1, -1, -1):
            for j in range(tlen -1, -1, -1):
                if s[i] == t[j]:
                    dp[i][j] = dp[i+1][j+1] + dp[i+1][j]
                else:
                    dp[i][j] = dp[i+1][j]

        return dp[0][0]
```
