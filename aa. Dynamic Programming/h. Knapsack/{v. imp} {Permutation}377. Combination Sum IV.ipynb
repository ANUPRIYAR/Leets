# Recursive Solution  => Here we need to find permutations not combinations, hence this will be a bit different

### TLE   

```python
from typing import List

class Solution:
    def combinationSum4(self, nums: List[int], target: int) -> int:
        return self.helper(nums, target)

    def helper(self, nums, target):
        if target == 0:
            return 1

        if target < 0:
            return 0

        result = 0
        for i in range(len(nums)):
            result += self.helper(nums,target - nums[i] )

        return result
```

# Top Down Solution


```python
class Solution:
    def combinationSum4(self, nums: List[int], target: int) -> int:
        dp = {}
        return self.helper(nums, target, dp)

    def helper(self, nums, target, dp):
        if target == 0:
            return 1

        if target < 0:
            return 0

        if target in dp:
            return dp[target]

        result = 0
        for num in nums:
            if num <= target:
                result += self.helper(nums, target - num, dp )
                dp[target] = result

        return result
```

# Bottom Up 


```python
class Solution:
    def combinationSum4(self, nums: List[int], target: int) -> int:
        dp = [0 for _ in range(target + 1)]
        dp[0] = 1

        for t in range(1, target + 1):
            for i in range(len(nums)):
                if nums[i] <= t:
                    dp[t] += dp[t-nums[i]]

        return dp[target]
```
