Input: n = 1  
Output: 1  
 

Constraints:  

1 <= n <= 19  


## Recursive Top Down


```python
class Solution:
    def numTrees(self, n: int) -> int:
        dp = [-1 for _ in range(n +1)]
        return self.dfs(dp, n)

    def dfs(self, dp, n):
        if n <= 1:
            return 1

        if dp[n] == -1:
            ans = 0
            for i in range(1, n+1):
                ans += self.dfs(dp, i-1) * self.dfs(dp, n-i)
            dp[n] = ans
        return dp[n]
```

## Bottom Up 


```python
#Since for recursive solution we are starting from n , In Bottom up , we take (1 to n + 1)
class Solution:
    def numTrees(self, n: int) -> int:
        if n ==0 or n ==1:
            return 1
            
        dp = [0 for _ in range(n +1)]
        dp[0]= 1

        ans = 0
        for i in range(1, n + 1):
            for j in range(i):
                dp[i] += dp[j] * dp[i-1-j]

        return dp[n]
```
