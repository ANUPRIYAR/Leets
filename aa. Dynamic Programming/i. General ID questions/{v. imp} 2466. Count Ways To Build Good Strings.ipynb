# Recursion with top down 


```python
class Solution:
    def countGoodStrings(self, low: int, high: int, zero: int, one: int) -> int:
        dp = {}
        count = self.helper(low, high, zero, one, 0, dp)
        return count % (10**9 + 7)

    def helper(self, low, high, zero, one, length, dp):
        count = 0
        if length > high:
            return 0

        if low <= length <= high:
            count += 1

        if length not in dp:
            add_zero = self.helper(low, high, zero, one, length + zero, dp)
            add_one =  self.helper(low, high, zero, one, length + one, dp)
            count += add_zero + add_one
            dp[length] = count
        return dp[length]
```

# Bottom up


```python
class Solution:
    def countGoodStrings(self, low: int, high: int, zero: int, one: int) -> int:
        dp = [0] * (high + 1)
        dp[0] = 1

        for i in range(1, high + 1):
            dp[i] = dp[i - zero] + dp[i - one ]        

        return sum(dp[low:high +1])% (10**9 + 7)
```
