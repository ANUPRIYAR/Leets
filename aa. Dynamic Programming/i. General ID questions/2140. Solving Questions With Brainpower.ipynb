# Recursion with top down


```python
from typing import List
class Solution:
    def mostPoints(self, questions: List[List[int]]) -> int:
        dp = {}
        return self.helper(questions, 0, dp)

    def helper(self, ques, idx, dp):
        if idx > len(ques) - 1:
            return 0

        if idx not in dp:
            attempt = ques[idx][0] + self.helper(ques, idx + ques[idx][1] + 1 , dp)
            skip = self.helper(ques,  idx + 1, dp)
            dp[idx]  = max(attempt, skip)

        return dp[idx] 
```

# Bottom up using Hashamp


```python
class Solution:
    def mostPoints(self, questions: List[List[int]]) -> int:
        n = len(questions)
        dp = {}
        for i in range(n-1, -1, -1):
            next_index = i + questions[i][1] + 1
            attempt = questions[i][0] + dp.get(next_index, 0)
            skip = dp.get(i+1, 0)
            dp[i] = max(attempt, skip)
        return dp[0]
```
