## Problem Statement
<p> Given an array arr, you transform it daily based on the following rules: </p>
<li>
 If an element is smaller than both its left and right neighbors, increment it by 1. 
    </li>
    <li>
If an element is larger than both its left and right neighbors, decrement it by 1.
        </li>
<p>The first and last elements of the array do not change.
Continue transforming the array until no more changes occur. Return the final state of the array.</p>

<ul>
<strong> Examples </strong>
    <li>
Example 1
        </li>
    <li>
Input: arr = [5, 2, 8, 1, 6]
        </li> 
    <li>
Expected Output: [5, 5, 5, 5, 6]
         </li> 
    <li>
Justification:
        </li>
<p> On the first day, 2 is incremented to 3, 8 is decremented to 7, and 1 is incremented to 2. The first and last elements remain unchanged. The final state of the array after the first day is [5, 3, 7, 2, 6].
After second day: [5, 4, 6, 3, 6].
After third day: [5, 5, 5, 4, 6].
After forth day: [5, 5, 5, 5, 6].</p>
</ul>

Constraints:
<li>
3 <= arr.length <= 100
    </li>
    <li>
1 <= arr[i] <= 100
</li>

```python
class Solution:
    def transformArray(self, arr):
        changed = True
        while changed:
            changed = False
            for i in range(1, len(arr) - 1):
                if arr[i-1] > arr[i] < arr[i+1]:
                    arr[i] += 1
                    changed = True
                elif arr[i-1] < arr[i] > arr[i+1]:
                    arr[i] -= 1
                    changed = True
        return arr
```

## Another implemntation : Same logic 

```python
class Solution:
    def transformArray(self, arr):        
        changed = True
        
        # Continue transforming until no changes occur
        while changed:
            changed = False
            # Store the first three elements for comparison
            prev, curr, next = arr[0], arr[1], arr[2]
            
            # Loop through the array starting from the second element to the second last element
            for i in range(1, len(arr) - 1):
                # If current element is less than both its neighbors, increment it
                if curr < prev and curr < next:
                    arr[i] += 1
                    changed = True
                # If current element is greater than both its neighbors, decrement it
                elif curr > prev and curr > next:
                    arr[i] -= 1
                    changed = True
                # Move to the next set of elements
                if i < len(arr) - 2:
                    prev, curr, next = curr, next, arr[i + 2]
        
        return arr
```

```python

# Test the method with different inputs
sol = Solution()
arr1 = [5, 2, 8, 1, 6]
arr2 = [3, 7, 1, 5, 2]
arr3 = [9, 9, 9, 9, 9]

print(sol.transformArray(arr1))  # [5, 5, 5, 5, 6]
print(sol.transformArray(arr2))  # [3, 3, 3, 3, 2]
print(sol.transformArray(arr3))  # [9, 9, 9, 9, 9]
```
