## Minimum Platforms
<strong>Difficulty: Medium </strong> 
<p>Given arrival and departure times of all trains that reach a railway station. Find the minimum number of platforms required for the railway station so that no train is kept waiting.<p/>
<p> Consider that all the trains arrive on the same day and leave on the same day. Arrival and departure time can never be the same for a train but we can have arrival time of one train equal to departure time of the other. At any given instance of time, same platform can not be used for both departure of a train and arrival of another train. In such cases, we need different platforms.</p>

<p>Note: Time intervals are in the 24-hour format(HHMM) , where the first two characters represent hour (between 00 to 23 ) and the last two characters represent minutes (this will be <= 59 and >= 0).</p>

<p> Examples:</p>

<li>
<strong> Input:</strong> arr[] = [0900, 0940, 0950, 1100, 1500, 1800], 
            dep[] = [0910, 1200, 1120, 1130, 1900, 2000]
    </li>
    <li>
Output: 3
        </li>
        <li>
<strong> Explanation: </strong>There are three trains during the time 0940 to 1200. So we need minimum 3 platforms.
            </li>

<li>
Input: arr[] = [0900, 1235, 1100], 
            dep[] = [1000, 1240, 1200]
 </li> 
 <li>
Output: 1
      </li> 
      <li>
Explanation: All train times are mutually exlusive. So we need only one platform
          </li>
 <li>         
<strong>Input:</strong> arr[] = [1000, 0935, 1100], 
            dep[] = [1200, 1240, 1130]
     </li>
      <li>  
<strong>Output:</strong> 3
          </li>
<strong>Explanation:</strong All 3 trains have to be their from 11:00 to 11:30

<li>
Expected Time Complexity: O(nLogn)
    </li>
    <li>
Expected Auxiliary Space: O(n)
</li>

<li>
Constraints:
</li>
<li>
1≤ number of trains ≤ 50000
0000 ≤ arr[i] ≤ dep[i] ≤ 2359
</li>


```python
class Solution:    
    #Function to find the minimum number of platforms required at the
    #railway station such that no train waits.
    def minimumPlatform(self,arr,dep):
        arr.sort()
        dep.sort()
        
        ans = 1
        count = 1
        i = 1
        j = 0
        max_platforms = 1
        
        while i < len(arr) and j < len(dep):
            if arr[i] <= dep[j]:
                count += 1
                i += 1
            else:
                count -= 1
                j += 1
                
            max_platforms = max(max_platforms, count)
                
        return max_platforms
        
```
