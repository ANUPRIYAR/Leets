## Job Sequencing Problem
<h3> Medium</h3> Accuracy: 34.51%Submissions: 251K+Points: 4

<p></p> Given a set of n jobs where each jobi has a deadline and profit associated with it.
 Each job takes 1 unit of time to complete and only one job can be scheduled at a time. We earn the profit associated with a job if and only if the job is completed by its deadline.
Find the number of jobs done and the maximum profit. </p>

Note: Jobs will be given in the form (Jobid, Deadline, Profit) associated with that Job. Deadline of the job is the time on or before which job needs to be completed to earn the profit. </strong></pre>

<p><strong class="example">Examples :</strong></p>

<pre>
<strong>Input: </strong> Jobs = [[1,4,20],[2,1,1],[3,1,40],[4,1,30]]`
<strong>Output:</strong> 2 60
<strong>Explanation:</strong> Job1 and Job3 can be done with maximum profit of 60 (20+40).
 </pre>

<pre>
<strong>Input: </strong>Jobs = [[1,2,100],[2,1,19],[3,2,27],[4,1,25],[5,1,15]]
<strong>Output:</strong> 2 127
<strong>Explanation:</strong> 2 jobs can be done with maximum profit of 127 (100+27).
 </pre>
Expected Time Complexity: O(nlogn)
Expected Auxilliary Space: O(n)


<li>
Constraints:
</li>
<li>
1 <= n <= 105
</li>
<li>
1 <= Deadline,id <= n
</li>
    <li>
1 <= Profit <= 500
</li>

```python

class Job:
    
    # Job class which stores profit and deadline.
    
    def __init__(self,profit=0,deadline=0):
        self.profit = profit
        self.deadline = deadline
        self.id = 0
      

class Solution:
    
    #Function to find the maximum profit and the number of jobs done.
    def JobScheduling(self,Jobs,n):
        last_deadline = 0
        for j in Jobs:
            last_deadline = max(last_deadline, j.deadline)
            
        Jobs.sort(key = lambda x: x.profit, reverse =True)
        
        job_array = [-1]* (last_deadline + 1)
        total_profit = 0
        
        for j in Jobs:
            deadline = j.deadline
            id = j.id
            profit = j.profit

            while job_array[deadline] != -1 and deadline > 1:
                deadline -= 1
                
            if deadline == 1 and job_array[deadline] != -1:
                continue
            job_array[deadline] = id
            total_profit += profit
        
            
        num_jobs = 0
        for i in range(len(job_array)):
            if job_array[i] != -1:
                num_jobs += 1
            
            
        return num_jobs, total_profit
```
