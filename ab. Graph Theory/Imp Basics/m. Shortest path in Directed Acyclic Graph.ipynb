## Using Topo Sort and DFS

```python
#User function Template for python3
from collections import defaultdict
from typing import List
import math

class Solution:
    def shortestPath(self, n : int, m : int, edges : List[List[int]]) -> List[int]:\
        
        def toposort(node, adj):
            vis[node]= 1      # Marked visited
            for neighbour in adj[node]:   # visit nieghbours
                nei = neighbour[0]        # 0th element is node
                if vis[nei] == 0:
                    toposort(nei, adj)                    
            stack.append(node)               # At the end of for loop, insert the node to stack
                        
        
        # Create adj list
        adj = defaultdict(list)
        for i in range(len(edges)):     
            u = edges[i][0]
            v = edges[i][1]
            wt = edges[i][2]
            adj[u].append([v, wt])    # On adj list, each node is appended as the pair of node and its weight
            
        # Iterate over the dfs and perform topo sort 
        vis = [0]*n
        stack = []    
        for i in range(n):
            if vis[i] == 0:                # Iterate over all the nodes and if not visited, perform toposort
                toposort(i, adj)
                
                
        # step 2 do the distance thing 
        dist = [math.inf]*n               # Intialize the distance list with infinity 
        for i in range(n):
            dist[0] = 0             # the start node will always be zero coz distance to itself
            
        while stack:
            node = stack.pop()          # pop the stack and visit the popped node's neighbours
            for neigh in adj[node]:
                v = neigh[0]
                wt = neigh[1] 
                
                if dist[node] + wt < dist[v]:        # then we check if dist stored at prev node index plus current weight is less than
                      dist[v] = dist[node] + wt        # dist stored at current index,if yes, we store the sum at current index    
                    
                    
        dist = [distance if distance != math.inf else -1 for distance in dist ]
        return dist
```
