<h2><a href="https://leetcode.com/problems/max-area-of-island">695. Max Area of Island</a></h2><h3>Medium</h3><hr><p>You are given an <code>m x n</code> binary matrix <code>grid</code>. An island is a group of <code>1</code>&#39;s (representing land) connected <strong>4-directionally</strong> (horizontal or vertical.) You may assume all four edges of the grid are surrounded by water.</p>

<p>The <strong>area</strong> of an island is the number of cells with a value <code>1</code> in the island.</p>

<p>Return <em>the maximum <strong>area</strong> of an island in </em><code>grid</code>. If there is no island, return <code>0</code>.</p>

<p><strong>Constraints:</strong></p>

<ul>
	<li><code>m == grid.length</code></li>
	<li><code>n == grid[i].length</code></li>
	<li><code>1 &lt;= m, n &lt;= 50</code></li>
	<li><code>grid[i][j]</code> is either <code>0</code> or <code>1</code>.</li>
</ul>


```python
from typing import List

class Solution:
    def maxAreaOfIsland(self, grid: List[List[int]]) -> int:
        rows, cols = len(grid), len(grid[0])        
        vis = [[0 for _ in range(cols)] for _ in range(rows)]
        directions = [(1, 0), (-1, 0), (0, 1), (0, -1)]

        def dfs(x, y, grid, vis):
            vis[x][y] = 1
            count = 1

            for dx, dy in directions:
                if 0 <= x + dx < rows  and 0 <= y + dy < cols and grid[x+dx][y+dy] == 1 and vis[x+dx][y+dy] == 0:
                    count += dfs(x+dx, y+dy, grid, vis) 

            return count

        areas = []
        for i in range(rows):
            for j in range(cols):
                if grid[i][j] == 1 and vis[i][j] == 0:
                    areas.append(dfs(i, j, grid, vis))


        return max(areas) if len(areas) > 0 else 0
```
