## 3127. Make a Square with the Same Color

You are given a 2D matrix grid of size 3 x 3 consisting only of characters 'B' and 'W'. Character 'W' represents the white color, and character 'B' represents the black color.

Your task is to change the color of at most one cell so that the matrix has a 2 x 2 square where all cells are of the same color.

Return true if it is possible to create a 2 x 2 square of the same color, otherwise, return false.

 

Example 1:







```python
from typing import List
class Solution:
    def canMakeSquare(self, grid: List[List[str]]) -> bool:      
        colormap = {'B':0 , 'W': 0}
        for i in range(2):
            for j in range(2):
                colormap[grid[i][j]] = colormap.get(grid[i][j], 0) + 1
                if colormap['B'] <= 2 or colormap['W'] <= 2:
                    if self.shift_matrix(grid, i, j):
                        return True
        return False


    def shift_matrix(self,grid,  i, j):
        colormap = {'B': 0 , 'W': 0}
        for si in range(2):
            for sj in range(2):
                colormap[grid[i+ si][j +sj]] = colormap.get(grid[i+ si][j +sj], 0) + 1
        if colormap['B'] > 2 or colormap['W'] > 2:
            return True
        return False
```

```python
grid = [["B","W","B"],["B","W","W"],["B","W","W"]]
sol = Solution()
sol.canMakeSquare(grid)
```

# Another Approach using directions

```python
from typing import List
class Solution:
    def canMakeSquare(self, grid: List[List[str]]) -> bool:    
        directions = [[0,0], [0,1], [1,0], [1, 1]]     # to traverse a 2D matrrix out of the given matrix 
        for i in range(2):
            for j in range(2):
                colormap = {'B': 0, 'W':0}
                for dx, dy in directions:
                    x, y = i + dx, j + dy
                    colormap[grid[x][y]] += 1
                    if colormap['B'] >= 3 or colormap['W'] >= 3:
                        return True
        return False
```

```python
grid = [["B","W","B"],["B","W","W"],["B","W","W"]]
sol = Solution()
sol.canMakeSquare(grid)
```
