## 3128. Right Triangles

Medium
You are given a 2D boolean matrix grid.
Return an integer that is the number of right triangles that can be made with the 3 elements of grid such that all of them have a value of 1.

Note:
A collection of 3 elements of grid is a right triangle if one of its elements is in the same row with another element and in the same column with the third element. The 3 elements do not have to be next to each other.
 
Example 1:





```python
from typing import List
class Solution:
    def numberOfRightTriangles(self, grid: List[List[int]]) -> int:
        rows, cols = len(grid) , len(grid[0])
        row_count = {}
        rowcount, colcount = [0]* rows, [0]* cols   # intialization for rowcount and colcount
        for i in range(rows):
            for j in range(cols):
                if grid[i][j] == 1:
                    rowcount[i] += 1   # to get row count
                    colcount[j] += 1  # to get col count
        triangles = 0
        for i in range(rows):
            for j in range(cols):
                if grid[i][j] == 1:
                    triangles += (rowcount[i] - 1)* (colcount[j]- 1)
        return triangles
```

```python
grid = [[0,1,0],[0,1,1],[0,1,0]]
sol = Solution()
sol.numberOfRightTriangles(grid)
```
