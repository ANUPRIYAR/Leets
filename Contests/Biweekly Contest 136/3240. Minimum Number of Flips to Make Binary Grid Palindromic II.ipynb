```python
class Solution:
    def minFlips(self, grid: List[List[int]]) -> int:
        rows, cols  = len(grid), len(grid[0])
        


        def adjustflips(totalones, flips, middleones, N):
            # if omes%4 !=0 and ones < 4 => flipsneeded = min(4-ones, ones
            rem = totalones % 4
            q = 1 + (totalones//4)
            if rem != 0:
                add = (4*q - totalones)
                print(f"add : {add}")
                remove = rem
                print(f"remove : {rem}")
                if add < remove:
                    if N - middleones >= add:
                        flips += add
                    else:
                        flips += remove
                else:
                    flips += remove

            return flips


        def FindTotalOnes(odd_flag, edgeones, gridsum):
            middleones = 0
            if odd_flag:
                middleones = gridsum - edgeones//2
                totalones = edgeones + middleones
                return totalones, middleones
            return edgeones, middleones


    
        rowedgesones , coledgesones , gridsum, middleones = 0, 0, 0, 0
        rowflips, colflips = 0, 0
        odd_row, odd_col = False, False

        # check cols => odd  => if yes dont count j == cols//2 -- not required
        if rows % 2 != 0:
            odd_row = True
        if cols % 2 != 0:
            odd_col = True


        # find rowflips and rowedgesones 
        for i in range(rows):
            for j in range(cols//2):
                if grid[i][j] != grid[i][cols -1 - i]:
                    print(f"grid[i][j] : {grid[i][j]}")
                    print(f"grid[i][cols -1 - i] : {grid[i][cols -1 - i]}")
                    rowflips += 1
                    rowedgesones += 2
                    print(f"rowedgesones : {rowedgesones}")
                else:
                    rowedgesones += grid[i][j] + grid[i][cols -1 - i]

        print(f"rowflips : {rowflips} rowedgesones : {rowedgesones}")

        # check rows => odd  => if yes subtract count i == rows//2 => anyways we are not traversing middle colyumn or row
        # find colflips and colsedgesones
        for j in range(cols):
            for i in range(rows//2):
                if grid[i][j] != grid[rows - 1 -j][j]:
                    colflips += 1
                    coledgesones = 2
                else:
                    coledgesones = grid[i][j] + grid[rows - 1 -j][j]

        print(f"colflips : {colflips} coledgesones : {coledgesones}")


        # find gridsum
        for row in grid:
            gridsum += sum(row)

        print(f"gridsum : {gridsum}")
            

        # if rowflips < colflips 
        # run func to get total
        # if total % 4 != 0 rowflips >
        # if total < 4 => rowflips same 
        # if total % 4 not 0 , then if middleones == total % 4 , rowflips+= min(middleones , rows - middle ones ) and viceversa 
        # if rowflips < colflips:
        totalones_r, middleones_r = FindTotalOnes(odd_col, rowedgesones, gridsum)
        adjusted_rowflips = adjustflips(totalones_r, rowflips, middleones_r, rows)
        print(f"totalones_r : {totalones_r} middleones_r : {middleones_r} ")
        print(f"adjusted_rowflips : {adjusted_rowflips}")

        totalones, middleones = FindTotalOnes(odd_row, coledgesones , gridsum)
        adjusted_colflips = adjustflips(totalones, colflips, middleones, cols)
        print(f"totalones : {totalones} middleones : {middleones} ")
        print(f"adjusted_colflips : {adjusted_colflips}")

        ones = max(totalones, totalones_r)
        
        # if  rows == 1 or cols == 1:
        #     return max(adjusted_rowflips, adjusted_colflips)
        # else:
        #     return min(adjusted_rowflips, adjusted_colflips)
        return min(adjusted_rowflips, adjusted_colflips)


        # func : 
        # odd = True , middle = n// 2  
        # middle ones => gridsum - edgeones//2
        # total = edgeones + middleones
```
