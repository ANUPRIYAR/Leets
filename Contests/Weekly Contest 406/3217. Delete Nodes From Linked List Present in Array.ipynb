<p><strong>Constraints:</strong></p>

<ul>
	<li><code>1 &lt;= nums.length &lt;= 10<sup>5</sup></code></li>
	<li><code>1 &lt;= nums[i] &lt;= 10<sup>5</sup></code></li>
	<li>All elements in <code>nums</code> are unique.</li>
	<li>The number of nodes in the given list is in the range <code>[1, 10<sup>5</sup>]</code>.</li>
	<li><code>1 &lt;= Node.val &lt;= 10<sup>5</sup></code></li>
	<li>The input is generated such that there is at least one node in the linked list that has a value not present in <code>nums</code>.</li>
</ul>


```python
from typing import List, Optional

# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
class Solution:
    def modifiedList(self, nums: List[int], head: Optional[ListNode]) -> Optional[ListNode]:
        current = head
        dummy = ListNode()
        dumm = dummy
        
        nums = set(nums)  # To remove duplicates

        while current:
            if current.val in nums:
                value = current.val
                while current and current.val == value:  #  skipping same value
                    next = current.next             # we get the next element 
                    current = next                 # then assign current to next
                
            else:
                dummy.next = current              # if condition satifies , assign an empty linked list to the current linked list
                dummy = dummy.next       # move both current and dummy node forward
                current = current.next
        
        dummy.next = None            # At the end of loop assign dummy.next to None
        return dumm.next
```
